# Position change start vs finish

```python
print(df[['Driver_Name','Pos_Change']])
```

```python
# gather data for graph

data = df[['Driver_Name', 'Start_Pos', 'Pos_Change']]
data = data.sort_values('Start_Pos')
name = data['Driver_Name']
pos_change = data['Pos_Change']

# plot position change
fig, ax = plt.subplots(figsize=(12,6))
clrs = ['red' if (x < 0) else 'green' for x in pos_change ]
ax.bar(name, pos_change, color=clrs, zorder=3)
ax.set_title('Position Change Start vs Finish -- Food City 500')
ax.set_xlabel('Driver')
ax.set_xticks(name)
ax.set_xticklabels(name, rotation=90, ha='center')
ax.set_ylabel('Position Change')
ax.set_yticks(np.arange(min(pos_change), max(pos_change)+1, 5))
ax.set_facecolor('gray')
fig.set_facecolor('darkgray')
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='-')
plt.grid(axis='y', linestyle='--', linewidth=0.5, zorder=0)
#plt.savefig('Position_change_start_vs_finish__food_city_500.png', bbox_inches = 'tight')
plt.show()
```
